```python
def get_block(img, dim, pos):
    slice_ = int(img.shape[0] / 4)
    line = pos // dim 
    col = int(pos % dim)
    return img[line*slice_:(line+1)*slice_,col*slice_:(col+1)*slice_]
```

```python
seq_l1 = [5,10,12,2]
seq_l2 = [7,15,0,8]
seq_l3 = [11,13,1,6]
seq_l4 = [3,14,9,4]

l1 = get_block(baboon, 4, seq_l1[0])
for pos in range(1,len(seq_l1)):
    b = get_block(baboon,4, seq_l1[pos])
    l1 = np.concatenate((l1,b),axis=1)
    
l2 = get_block(baboon, 4, seq_l2[0])
for pos in range(1,len(seq_l2)):
    b = get_block(baboon,4, seq_l2[pos])
    l2 = np.concatenate((l2,b),axis=1)

l3 = get_block(baboon, 4, seq_l3[0])
for pos in range(1,len(seq_l3)):
    b = get_block(baboon,4, seq_l3[pos])
    l3 = np.concatenate((l3,b),axis=1)
    
l4 = get_block(baboon, 4, seq_l4[0])
for pos in range(1,len(seq_l4)):
    b = get_block(baboon,4, seq_l4[pos])
    l4 = np.concatenate((l4,b),axis=1)
```

```python
mos = np.concatenate((l1,l2,l3,l4),axis=0)
```

```python
plt.imshow(mos,cmap='gray')
```
